```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#Overwatch Stats


```python
phs_2018_1 = pd.read_csv('phs_2018_stage_1.csv')
phs_2018_2 = pd.read_csv('phs_2018_stage_2.csv')
phs_2018_3 = pd.read_csv('phs_2018_stage_3.csv')
phs_2018_4 = pd.read_csv('phs_2018_stage_4.csv')
phs_2018_playoffs = pd.read_csv('phs_2018_playoffs.csv')
phs_2019_1 = pd.read_csv('phs_2019_stage_1.csv')
phs_2019_2 = pd.read_csv('phs_2019_stage_2.csv')
phs_2019_3 = pd.read_csv('phs_2019_stage_3.csv')
phs_2019_4 = pd.read_csv('phs_2019_stage_4.csv')
phs_2019_playoffs = pd.read_csv('phs_2019_playoffs.csv')
phs_2020_1 = pd.read_csv('phs_2020_1.csv')

all = [phs_2018_1, phs_2018_2, phs_2018_3, phs_2018_4, phs_2018_playoffs, phs_2019_1, phs_2019_2, phs_2019_3, phs_2019_4, phs_2019_playoffs, phs_2020_1]
allGames = pd.concat(all)
```

#Average Total Healing Dealt by OW healers

```python
#Set size of the figure
plt.figure(figsize = (7,8))

#Charicter variables containing stat amount of healing dealt
mercy = allGames.loc[(allGames['hero'] == 'Mercy') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
moira = allGames.loc[(allGames['hero'] == 'Moira') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
baptiste = allGames.loc[(allGames['hero'] == 'Baptiste') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
brigitte = allGames.loc[(allGames['hero'] == 'Brigitte') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
lucio = allGames.loc[(allGames['hero'] == 'Lúcio') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
ana = allGames.loc[(allGames['hero'] == 'Ana') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]
zenyatta = allGames.loc[(allGames['hero'] == 'Zenyatta') & (allGames['stat_name'] == 'Healing Done'), ['stat_amount']]

#labels on x axis per healer
labels = ['Mercy', 'Moira', 'Baptiste', 'Brigitte', 'Lúcio', 'Ana', 'Zenyatta']

#cat whiskers boxplot including all needed variables to display graph
boxes = plt.boxplot([mercy, moira, baptiste, brigitte, lucio, ana, zenyatta], labels=labels, patch_artist=True, medianprops={'linewidth':2}, showfliers=False)

#Set box styles
for box in boxes['boxes']:
  #Set edge color
  box.set(color='#4286f4', linewidth=2)

  #Change fill color 
  box.set(facecolor= '#e0e0e0')

#Header and y axis label
plt.title('Average Total Healing Done within Overwatch League 2018-2020', fontdict={'fontweight':'bold'})
plt.ylabel('Healing Done', fontdict={'fontweight':'bold'})

#save graph as png
plt.savefig('Overwatch_total_healing.png', dpi=300)

#display the graph
plt.show()
```
